```python
imdb_dataset = requests.get('https://www.imdb.com/search/title/?sort=num_votes,desc&start=1&title_type=feature&year=1950,2012')
```

```python
soup = BeautifulSoup(imdb_dataset.text, 'html5lib')

content = soup.find("div", id = "main")

movies = content.findAll("div", class_ = "lister-item-content")

print("Total Number of Movies: ",len(movies))
```

```python
data=[]
for movie in movies:
    name = movie.find("a")
    year = eval(name.next_element.next_element.next_element.text)
    name = name.text
    runtime = int(movie.find("span", class_="runtime").text.split(" ")[0])
    # print(runtime)

    data.append((name,year,runtime))
```

```python
for entry in data:
    print(entry)
```

```python
from sklearn import datasets


boston_data = datasets.load_boston()
house_data = pd.DataFrame(data=boston_data["data"], columns=boston_data["feature_names"])

house_data.head()
```

```python
sns.displot(house_data["DIS"], kde = True)
plt.xlabel("DIS", fontsize = 10)
plt.ylabel("Count", fontsize = 10)
plt.title("Univariate outlier graph", fontsize = 12)
plt.show()
```

```python
sns.boxplot(house_data[["RM","DIS","LSTAT"]])
plt.show()
```

```python
limit = 3
mean = house_data["DIS"].mean()
dev = house_data["DIS"].std()

house_data["z"] = (house_data["DIS"] - mean)/dev

outliers = house_data["z"][(house_data["z"] > 3) | (house_data["z"] < -3)]
simple_vals = house_data["z"][(house_data["z"] < 3) & (house_data["z"] > -3)]
```

```python
plt.scatter(x = [0]*len(simple_vals),y = simple_vals, s=0.6)
plt.scatter(x = [0]*len(outliers),y = outliers, s=11, c="red")
plt.show()
```

```python
country_data = pd.read_csv("./country.csv")
country_data.head()
```

```python
import numpy as np

column = "Area (sq. mi.)"

lower_q = np.percentile(country_data[column], 5, method = 'midpoint') # 5 percentile as lower bound
upper_q = np.percentile(country_data[column], 95, method = 'midpoint') # 95 pecentile as upper bound

IQR = upper_q - lower_q

upper = upper_q + 1.5*IQR
lower = lower_q - 1.5*IQR

outliers = country_data[column][(country_data[column] < lower) | (country_data[column] > upper)]

outliers
```

```python
plt.scatter(x = [0]*len(country_data[column]),y = country_data[column], s=0.6)
plt.scatter(x = [0]*len(outliers),y = outliers, s=20, c="green")
plt.title("Z values of Area")
plt.show()
```

```python
print(len(country_data)) # length before deleting outliers
country_data.drop(outliers.index, inplace=True)
print(len(country_data)) # length after
```
