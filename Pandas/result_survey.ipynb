### Filtering

```python
filt = (df["YearsCode"] > '5')  & (df["YearsCodePro"] > '5') & (df["Country"] == "United States of America")
```

```python
df.loc[filt]
```

```python
high_salary = df["CompTotal"] > 1000000
```

### NOTE: .sort_values(by = ["Column Name"]) to sort values according to the specified sorted column

```python
df.loc[high_salary,["Country","LanguageHaveWorkedWith"]].sort_values(by = ["Country"])
```

isin() to be used when to filter out more values of a column

```python
countries = ["India", "Germany", "France", "Canada", "United Kingdom"]
filt = df["Country"].isin(countries)
```

```python
# df.loc[filt]
df.loc[filt, "Country"]
```
