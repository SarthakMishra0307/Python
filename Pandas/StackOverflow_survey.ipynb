### Grouping

It involes, splitting the object, applying the function and combining the result

```python
df['Country'].value_counts()
```

```python
gb = df.groupby(['Country'])

# this object contains a bunch of groups
```

```python
gb.get_group('India')
```

This would be similar to running a filter on our original dataframe

```python
filt = df['Country'] == 'India'

df[filt]
```

But instead of grabbing the result for one country, GroupBy instead splits all these responses up by country name

```python
df[filt]['Age'].value_counts()
```

```python
df[filt]['Gender'].value_counts()
```

again we did the same thing but using groupby it will show a 'series' of all the countries in the same format

```python
gb['Age'].value_counts().head(50)

# use loc for finding the value count for a specific country(value) 
# use ncan use ormalize also
# returns a series with multiple index
```

Performing aggregrate functions on the groupby variable

```python
filt = df['Country'] == 'India'

df[filt]['LanguageHaveWorkedWith'].str.contains('Python')
#  .sum
```

Applying the same thing on group by object will throw an error because it is no longer a series, it is now an group by object, so to resolve this we will use the apply() method to use this command

```python
# gb.['LanguageHaveWorkedWith'].str.contains('Python').sum()
```

```python
gb['LanguageHaveWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
# x['India']
# lambda will be a series
```

```python
country_respondents = df['Country'].value_counts()
country_respondents
```

```python
country_uses_python = gb['LanguageHaveWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_uses_python
```

We can combine more than one series in pandas using concat function

axis = columns matches the indexes of data

```python
python_df = pd.concat([country_respondents,country_uses_python],axis= 'columns',sort= False)

python_df
```

```python
python_df.rename(columns={'Country' : 'Respondents', 'LanguageHaveWorkedWith':'WhoKnowsPython'}, inplace=True)
```

```python
python_df['PctKnowsPython'] = (python_df['WhoKnowsPython']/python_df['Respondents'])*100
python_df
```

```python
python_df.sort_values(by= 'PctKnowsPython', ascending= False, inplace= True)
python_df.head(50)
```
