```python
student = {
    "Name": ["Ankit", "Ashwani","Akshansh", "Gaurav", "Krish","Sarthak", "Swakshwar"],
    "Section": ["CH", "CH", "MP", "CH", "SS", "MP", "MP"],
    "Registration_Id": [12018329, 12016043, 12015923, 12014917, 12018275, 12018433, 12015939],
    # "Registration_Id": ["12018329", "12016043", "12015923", "12014917", "12018275", "12018433", "12015939"],
    "City" : ['Fatehabad', "Bahadurgarh", "Gurgaon", "Lucknow", "Jaipur", "Ghaziabad", "Kolkata"],
    "State" : ["Harayana", "Harayana", "Harayana", "Uttar Pradesh", "Rajasthan" ,"Uttar Pradesh" ,"Kolkata"]
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(student)
```

```python
df
```

Adding Column

```python
df.rename(columns={'Name':'First'},inplace=True) 
```

```python
# Just make a list of the new values and assign the name when calling the column

last =['Kumar','Ahlawat','Garg','Patel','Yadav','Mishra','Ghosh']
df['Last'] = last
df
```

Combining columns

```python
df['Full_Name'] = df['First'] + ' ' +df['Last']
df
```

Removing Columns

```python
df.drop(columns={'First', 'Last'}, inplace= True)
# first preview and if you are sastisfied with the changes, use inplace attribute
```

```python
df['Full_Name'].str.split(' ',expand = True)
```

If we want to assign those values to different columns we can use the expand the columns

```python
df[['First', 'Last']] = df['Full_Name'].str.split(' ',expand = True)
```

```python
df
```

Dropping columns

```python
df.drop(columns= 'Full_Name', inplace=True)
```
